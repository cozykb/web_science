### 1.2
Normalize the scores from the 2 models (by subtracting the mean and dividing with the standard deviation) and compute a combined rank using the Weighted Sum method with $\alpha=\beta=1$.

```python
import copy
def pd_norm(df:pd.DataFrame, *model_list):
    df=copy.deepcopy(df)
    for model in model_list:
        df[model] = (df[model]-df[model].mean())/df[model].std()
    return df

def weight_sum(df:pd.DataFrame, model_list:list, weight_list:list):
    df=copy.deepcopy(df)
    df['weight_score'] = 0
    for model, weight in zip(model_list,weight_list):
        df['weight_score'] += weight*df[model]
    df = df.sort_values(by=['weight_score'],ascending=False)
    return df
```

### 1.3
Merge the ranking from the 2 models using the Borda Fuse method.

```python
def round_robin(df:pd.DataFrame, model_list:list, rank:int=5):
    all_rank_list = []
    output_list = []
    for model in model_list:
        all_rank_list.append(list(df.sort_values(by=[model],ascending=False)['item_id']))
    while len(output_list) != rank:
        for i in range(len(all_rank_list[0])):
            for rank_list in all_rank_list:
                if rank_list[i] not in output_list:
                    output_list.append(rank_list[i])
    return output_list

def borda_fuse(df:pd.DataFrame, model_list:list, rank:int=5):
    df=copy.deepcopy(df)
    all_rank_list = []
    df['bf_score'] = 0
    score = range(rank-1,-1,-1)
    for model in model_list:
        all_rank_list.append(list(df.sort_values(by=[model],ascending=False).index))
    for i in range(rank):
        for rank_list in all_rank_list:
            df.iloc[rank_list[i],-1] += score[i]
    return df

model_list = ['model1_score', 'model2_score']
```

```python
print(borda_fuse(df,model_list))
```

### 1.4
Merge the ranking from the 2 models using the Reciprocal Rank Fusion method with $k=0$.

```python
def reciprocal_rank_fusion(df:pd.DataFrame, model_list:list, rank:int=5, k:int=0):
    df=copy.deepcopy(df)
    all_rank_list = []
    df['rrf_score'] = 0
    for model in model_list:
        all_rank_list.append(list(df.sort_values(by=[model],ascending=False).index))
    for i in range(rank):
        for rank_list in all_rank_list:
            df.iloc[rank_list[i],-1] += 1.0/(k+i+1)
    return df
```

```python
print(reciprocal_rank_fusion(df,model_list))
```

```python
def cos_sim_indf(df:pd.DataFrame, uid:str):
    uid_vector = df.loc[uid,:].to_numpy()
    remain_uid_list = []
    remain_vector = []
    for id in df.index:
        if id != uid:
            remain_uid_list.append(id)
            remain_vector.append(df.loc[id,:].to_numpy())
    return remain_uid_list, cosine_similarity(remain_vector,[uid_vector])
```
